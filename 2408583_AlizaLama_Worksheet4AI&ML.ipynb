#Manual Gradient Calculation in Numpy

```python
import numpy as np
# Simple function f(x) = x^2
def f(x):
    return x ** 2
# Manual derivative (f’(x) = 2x)
def gradient(x):
    return 2 * x
# Update rule: x = x - learning_rate * gradient
x = 5.0
learning_rate = 0.1
for _ in range(10):  # Manually optimize for 10 steps
    x -= learning_rate * gradient(x)
    print("x:", x, "f(x):", f(x))
```

#Gradient Computations with Keras.

```python
import tensorflow as tf
x = tf.Variable(5.0)  # Trainable variable
with tf.GradientTape() as tape:
    y = x ** 2  # y = x^2
grad = tape.gradient(y, x)  # Computes dy/dx automatically
print(grad.numpy())  # Output: 10.0
```

#Task 1: Data Preparation

```python
import os
import numpy as np
from PIL import Image
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Paths
train_path = "/content/drive/MyDrive/AI&ML/DevanagariHandwrittenDigitDataset/Train"
test_path = "/content/drive/MyDrive/AI&ML/DevanagariHandwrittenDigitDataset/Test"

# Function to load data
def load_data(folder_path):
    images, labels = [], []

    class_names = sorted(os.listdir(folder_path))
    class_map = {name: idx for idx, name in enumerate(class_names)}

    for class_name in class_names:
        class_folder = os.path.join(folder_path, class_name)
        label = class_map[class_name]

        for file in os.listdir(class_folder):
            img_path = os.path.join(class_folder, file)

            img = Image.open(img_path).convert("L")   # grayscale
            img = img.resize((28, 28))                # resize
            img = np.array(img) / 255.0               # normalize

            images.append(img)
            labels.append(label)

    return np.array(images), np.array(labels)

# Load data
x_train, y_train = load_data(train_path)
x_test, y_test = load_data(test_path)

# Reshape
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Print shapes
print("Train:", x_train.shape, y_train.shape)
print("Test:", x_test.shape, y_test.shape)

# Show sample images
plt.figure(figsize=(10,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i].reshape(28,28), cmap='gray')
    plt.title(np.argmax(y_train[i]))
    plt.axis('off')
plt.show()
```

#Task 2: Build the FCN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Build FCN model
model = Sequential([
    Flatten(input_shape=(28, 28, 1)),

    Dense(64, activation='sigmoid'),
    Dense(128, activation='sigmoid'),
    Dense(256, activation='sigmoid'),

    Dense(10, activation='softmax')
])

# Show model summary
model.summary()
```

#Task 3: Compile the Model

```python
# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # because we used one-hot encoding
    metrics=['accuracy']
)

print("Model compiled successfully.")
```

#Task 4: Train the Model


```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# Callbacks
callbacks = [
    ModelCheckpoint("best_model.h5", save_best_only=True),
    EarlyStopping(monitor='val_loss', patience=3)
]

# Train model
history = model.fit(
    x_train, y_train,
    batch_size=128,
    epochs=20,
    validation_split=0.2,
    callbacks=callbacks
)
```

#Task 5: Evaluation

```python
# Evaluate model on test data
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

#Task 6: Save and Load the Model

```python
import tensorflow as tf
# Save the model
model.save("devnagari_model.h5")
print("Model saved.")
# Load the model
loaded_model = tf.keras.models.load_model("devnagari_model.h5")
print("Model loaded.")
# Re-evaluate loaded model
loss, accuracy = loaded_model.evaluate(x_test, y_test)
print("Loaded Model Loss:", loss)
print("Loaded Model Accuracy:", accuracy)
```
